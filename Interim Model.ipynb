```python
import requests
import pandas as p
import datetime
import json

stocks = json.load(open("stocks.json"))
Sectors = json.load(open("sectors.json"))['sectors']
```

```python
df = p.DataFrame()
for b in Sectors:
    print(b)
    for i in stocks["Stocks"][b]:
        #print(i)
        ab = requests.get(f"https://api.polygon.io/v2/aggs/ticker/{i}/range/1/day/2023-01-10/2023-03-03?adjusted=true&sort=asc&limit=50000&apiKey=el0FbXEICAXLDOh_2HHODn_QdqmBj_5WlhdNvp").json()
        try:
            fd = p.DataFrame(ab['results'])
            fd['Ticker'] = i
            date = []
            for i in fd['t']:
                f = datetime.datetime.fromtimestamp(i/1000).strftime('%Y-%m-%d')
                date.append(f)
            df = df.append(fd, ignore_index=True)
            fd['Time'] = date
        except KeyError: 
            continue
```

```python
df.to_csv('Test Set.csv', index=False)
```
